## Variants

```python
x = np.array([1,2,3,4,5])
y = np.array([2,6,76,32,21])
plt.plot(x, y)
plt.grid(True)
plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np.array([2,6,76,32,21])
plt.plot(x, y)
plt.xscale('linear')
plt.yscale('log')
plt.show()
```

```python
labels = ["india", "USA", "UK", "Germany"]
values = [38, 32, 20, 10]
explode = (0.1, 0, 0, 0)
plt.pie(values, explode = explode, labels = labels, autopct='%1.1f%%', shadow = True, startangle = 90)
plt.axis('equal')
plt.show()
```

```python
mu = 100
sigma = 15
x = mu + sigma * np.random.randn(10**5)
n, bins, patches = plt.hist(x, 50, facecolor = 'red', alpha = 0.8, density=50)
plt.grid(True)
plt.show()
```

# Numpy

```python
dt = np.dtype(np.int32)
print(dt)
```

```python
# int8, int16, int32, int64 <==> i1, i2, i4, i8
for i in "1248":
    dt = np.dtype('i'+i)
    print(dt)
```

```python
# using endien notation
for i in '<>':
    dt = np.dtype(i+'i4')
    print(dt)
```

```python
dt = np.dtype([('age', np.int8)])
print(dt)
```

```python
dt = np.dtype([('age', np.int8)])
a = np.array([(10,),(20,),(30,)], dtype=dt)
print(a)
```

## Numpy Array

```python
a = np.array([1,2,3])
print(a)
```

```python
a = np.array([[1,2,3],[4,5,6]])
print(a)
```

```python
#minimum dimensions
a = np.array([1,2,3,4,5,6], ndmin=2)
print(a)
```

```python
# dtype parameter
a = np.array([[1,2,3],[4,5,6]], dtype=complex)
print(a)
```

```python
# ndarray.shape and reshaping
a = np.array([[1,2,3],[4,5,6]])
print(a)
print(a.shape)
a.shape = (3,2)
print(a)
```

```python
a = np.arange(24)
print(a)
print(a.ndim)
```

```python
b = a.reshape(2,4,3)
print(b)
```

```python
#numpy itemsize
a = np.array([1,2,3,4,5], dtype=np.int8)
print(a)
print(a.itemsize)

a = np.array([1,2,3,4,5], dtype=np.float32)
print(a)
print(a.itemsize)
```
